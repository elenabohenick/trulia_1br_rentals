```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# Check if url is scrapable

test_url = f"https://www.trulia.com/p/ny/manhattan/150-w-51st-st-1612-manhattan-ny-10019--2088845786/"
headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/39.0.2171.95 Safari/537.36'}
response = requests.get(test_url, headers=headers)
test_page = response.text
soup = BeautifulSoup(test_page,"html5lib")
print(soup.prettify())
```

```python
# create a search result page

def search_page(page_num):
    url = f"https://www.trulia.com/for_rent/New_York,NY/1_beds/{page_num}_p/"
    headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/39.0.2171.95 Safari/537.36'}
    response = requests.get(url, headers=headers)
    search_page = response.text
    soup = BeautifulSoup(search_page,"html5lib")
    return(soup)
```

```python
# get a list of html elemnts containing unique apt listing links

def get_link_elements(page_count):
    elements = []

    for i in range(1,page_count+1):
        element = search_page(i).find_all(class_='tileLink phm')
        elements.extend(element)
    
    return(elements)
```

```python
elements = get_link_elements(122)
len(elements)
```

```python
# get a list of apt listing links and zip code list

link_list = []

for element in elements:
    apt_link = str(element).split('"')
    parsed_link = apt_link[7].split('/')[-1]

    full_link = "https://www.trulia.com/p/ny/manhattan/" + parsed_link
    link_list.append(full_link)
```

```python
links_file = open('trulia_rentals/trulia_apt_links.txt', 'w')

for item in link_list:
    links_file.write("%s\n" % item)
```

```python
links_file = open('trulia_rentals/trulia_apt_links.csv', 'w')

for item in link_list:
    links_file.write("%s\n" % item)
```
